### QUANC8
QUANC8 - адаптивный численный метод для вычисления определённых интегралов. Он основан на формуле Ньютона-Котеса 8-го порядка (9 узлов) и автоматически подстраивает шаг разбиения, чтобы достичь заданной точности.   
Формула Ньютона-Котеса 8-го порядка (9 узлов) (Milton Abramowitz and Irene A. Stegun, 1964), из которой мы берем веса:   
![Формула Ньютона-Котеса 8-го порядка](./meta/2.png)   
Поскольку по условию дан интервал [3, 5] и шаг h=0.25 первый множитель в формуле запишем как 8h/28350.   
QUANC8 реализован в программе как функция quanc8(f, a, b, tol=1e-8, max_depth=30)   
Входные параметры:   
    f – функция, которую нужно проинтегрировать;   
    a и b – границы интегрирования;   
    tol – требуемая точность (допустимая погрешность) по умолчанию 1e-8 (10^-8);   
    max_depth - ограничение на число разбиений, по умолчанию 30.   
Возвращает суммарный интеграл.
